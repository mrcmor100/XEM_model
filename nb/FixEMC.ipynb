```python
def calc_nprat(x_arr, q2_arr):
    xlo, xhi, q2lo, q2hi, deltax, deltaq2 = 0, 0, 0, 0, 0, 0
    A1, A2, A3, A4, A12, A34, A1234       = 0, 0, 0, 0, 0, 0, 0
    f2r_arr = []
    
    for n in range(len(x_arr)):
        x = x_arr[n]
        q2 = q2_arr[n]
        for i in range(0,99): #Loop Over X
            q2lo=0.0
            q2hi=0.0
            xlo=0.0
            xhi=0.0

            if((x >= xtab[i]) and (x < xtab[i+1])):
                xhi=xtab[i+1]
                xlo=xtab[i]
                deltax=xhi-xlo
            
                for j in range(0,9):  #Loop over Q2
                    if((q2 >= q2tab[j]) and (q2 < q2tab[j+1])):
                        q2hi=q2tab[j+1]
                        q2lo=q2tab[j]
                        deltaq2=q2hi-q2lo
                        #print(i,j)
                        A1=nonp[j][i]
                        A2=nonp[j][i+1]
                        A3=nonp[j+1][i]
                        A4=nonp[j+1][i+1]

                        A12=(A1*(xhi-x) + A2*(x-xlo))/deltax
                        A34=(A3*(xhi-x) + A4*(x-xlo))/deltax

                        A1234=(A12*(q2hi-q2)+A34*(q2-q2lo))/deltaq2
                    #End Q2 check if
                #End Q2 Loop
            #End X check if
        #End X Loop
        f2rat=A1234
        #Apply John's scale factor
        f2rat=(1.0-0.013*np.log(q2/16.0)) * (1.0+f2rat)-1.0
        #print(f2rat)
    
        f2r_arr.append(f2rat)
    #End of point in X,Q2 pair
    #print('leaving calc_nprat ',f2rat)
    return f2r_arr

def calc_iso(A,Z,X,Q2):
    np_ratio = calc_nprat(X,Q2)
    iso_arr = []
    for i in range(len(np_ratio)):
        iso_arr.append(0.5*(1+np_ratio[i]) / (1/A*(Z+(A-Z)*np_ratio[i])))
    return iso_arr
```

```python
class plotEMC:
    def __init__(self, material):
        self.material = material
        self.E = 10.602
        if(self.material=="deuterium"):
            self.A = 2
            self.Z = 1
            self.sym = "D"
            self.doIso = True
        elif(self.material=="helium3"):
            self.A = 3
            self.Z = 2
            self.sym = "He3"
            self.doIso = True
        elif(self.material=="helium4"):
            self.A = 4
            self.Z = 2
            self.sym = "He4"
            self.doIso = False
        elif(self.material=="beryllium"):
            self.A = 9
            self.Z = 4
            self.sym = "Be"
            self.doIso = True
        elif(self.material=="carbon"):
            self.A = 12
            self.Z = 6
            self.sym = "C"
            self.doIso = False
        elif(self.material=="aluminum"):
            self.A = 27
            self.Z = 13
            self.sym = "Al"
            self.doIso = True
        elif(self.material=="copper"):
            self.A = 64
            self.Z = 29
            self.sym = "Cu"
            self.doIso = True
        elif(self.material=="gold"):
            self.A = 197
            self.Z =79
            self.sym = "Au"
            self.doIso = True
        else:
            print("Error initializing Class! Unknown Material.")
        self.read_df()
            
    def get_material(self):
        print(self.material)
    
    def read_df(self):
        self.df = pd.read_csv('../output/%s_newFit.out' % self.material, delim_whitespace =True,names=['Y','A','Z','th','Ep','x_bj','sigIN_xem20','sigQE_xem20'], \
                    dtype={'Y':np.float64,'A':np.float64,'Z':np.float64,'th':np.float64,'Ep':np.float64,'x_bj':np.float64, \
                        'sigIN_xem':np.float64, 'sigQE_xem20':np.float64,'sigIN_xem20':np.float64})
        self.df['sigInXEM20'] = self.df['sigIN_xem20'] / 1000.
        self.df['sigQEXEM20'] = self.df['sigQE_xem20'] * 1000.
        self.df['sig_tot'] = self.df['sigInXEM20'] + self.df['sigQEXEM20']
        self.df['slacFit'] = self.emc_slac_fit(self.df['x_bj'].to_numpy(),self.A)

    def emc_slac_fit(self, x, A) :
        alpha = -0.070 + 2.189*x - 24.667*x**2 + 145.291*x**3 - 497.237*x**4 + 1013.129*x**5 - 1208.393*x**6 + 775.767*x**7 - 205.872*x**8
        C = np.exp(0.017 + 0.018*np.log(x) + 0.005*(np.log(x))**2)
        return C*A**alpha
            
    def plot_cs(self):
        fig2, ax2 = plt.subplots(1,1,figsize=(15, 1*7))
        self.read_df()
        ax2.set_title(self.material)
        self.df.plot.scatter('x_bj','sig_tot',xlim=(0.2,0.9),ax=ax2)
        
    def plot_ratio(self, plotEMC):
        fig1, ax1 = plt.subplots(1,1,figsize=(15, 1*7))
        self.read_df()
        ax1.set_title(self.material)
        self.df['ratio'] = (self.df['sig_tot'] / self.A) / (plotEMC.get_col('sig_tot') / plotEMC.get_A())
        self.df.plot.scatter('x_bj','ratio',color='blue',ax=ax1,label='No IsoScaler Correction')
        if(self.doIso==True):
            self.df['Q2'] = 4*self.E*self.df['Ep']*np.sin(self.df['th']*3.14159/360.)*np.sin(self.df['th']*3.14159/360.)
            self.df['isoCorr'] = calc_iso(self.get_A(), self.get_Z(), self.df['x_bj'].to_numpy(), self.df['Q2'].to_numpy())
            self.df['ratioIso'] = self.df['ratio'] * self.df['isoCorr']
            self.df.plot.scatter('x_bj','ratioIso',color='green',ax=ax1,label='IsoScaler Correction')
        self.df.plot('x_bj','slacFit',xlim=(0.2,0.98),ylim=(0.6,1.15),ax=ax1)
        ax1.set_ylabel("%s / %d   /   D / %d" % (self.sym, self.A, plotEMC.get_A()))
        ax1.set_xlabel(r'X')
    
    def plot_superratio(self, plotEMC):
        fig1, ax1 = plt.subplots(1,1,figsize=(15, 1*7))
        self.read_df()
        ax1.set_title(self.material)
        self.df['ratio'] = (self.df['sig_tot'] / self.A) / (plotEMC.get_col('sig_tot') / plotEMC.get_A())
        if(self.doIso==False):
            self.df['superRatio'] = self.df['slacFit'] / self.df['ratio']
            self.df.plot.scatter('x_bj','superRatio',color='blue',xlim=(0.2,0.9),ylim=(0.4,1.2),ax=ax1,label='No IsoScaler Correction Needed')
        elif(self.doIso==True):
            self.df['Q2'] = 4*self.E*self.df['Ep']*np.sin(self.df['th']*3.14159/360.)*np.sin(self.df['th']*3.14159/360.)
            self.df['isoCorr'] = calc_iso(self.get_A(), self.get_Z(), self.df['x_bj'].to_numpy(), self.df['Q2'].to_numpy())
            self.df['ratioIso'] = self.df['ratio'] * self.df['isoCorr']
            self.df['superRatio'] = self.df['slacFit'] / self.df['ratioIso']
            self.df.plot.scatter('x_bj','superRatio',color='green',xlim=(0.2,0.9),ylim=(0.7,1.3),ax=ax1,label='IsoScaler Corrected')
        #self.df.plot('x_bj','slacFit',xlim=(0.2,0.98),ylim=(0.6,1.15),ax=ax1)
        ax1.set_ylabel("(%s / %d   /   D / %d) / slacFit" % (self.sym, self.A, plotEMC.get_A()))
        ax1.set_xlabel(r'X')
    
    def get_poly_fit(self):
        coefs = poly.polyfit(self.df['x_bj'].to_numpy()[0:28],self.df['superRatio'].to_numpy()[0:28],5)
#emc = 0.92170 +1.7544*xtmp -3.7324*xtmp**2 - 0.24293*xtmp**3 + 6.7613*xtmp**4 - 4.6089*xtmp**5
        print("           emc = %7.5f + %7.5f*xtmp + %7.5f*xtmp**2 + \n	1        %7.5f*xtmp**3 + %7.5f*xtmp**4 + %7.5f*xtmp**5" % (coefs[0],coefs[1],coefs[2],coefs[3],coefs[4], coefs[5]))
    
    def get_col(self,column):
        return self.df[column].to_numpy()
    def get_A(self):
        return self.A
    def get_Z(self):
        return self.Z
    def get_doIso(self):
        print(self.doIso)
```
